# Big O
Big O notation is a way to describe the upper bound on the growth rate of an algorithm's time or space complexity in terms of the input size. It provides an asymptotic analysis of an algorithm's efficiency, focusing on the worst-case scenario. The notation is expressed as O(f(n)), where f(n) represents the growth rate of the algorithm concerning the input size (n).

image.pngHere are some common Big O notations and simple example codes:

1. **O(1) - Constant Time Complexity:**
   - The algorithm's runtime or space requirements remain constant regardless of the input size.

   ```python
   def constant_example(arr):
       return arr[0]
   ```

2. **O(log n) - Logarithmic Time Complexity:**
   - The algorithm's efficiency grows logarithmically with the input size.

   ```python
   def binary_search(arr, target):
       low, high = 0, len(arr) - 1
       while low <= high:
           mid = (low + high) // 2
           if arr[mid] == target:
               return mid
           elif arr[mid] < target:
               low = mid + 1
           else:
               high = mid - 1
       return -1
   ```

3. **O(n) - Linear Time Complexity:**
   - The algorithm's efficiency grows linearly with the input size.

   ```python
   def linear_example(arr):
       for item in arr:
           print(item)
   ```

4. **O(n log n) - Linearithmic Time Complexity:**
   - Common in efficient sorting algorithms like Merge Sort and Heap Sort.

   ```python
   def merge_sort(arr):
       if len(arr) <= 1:
           return arr
       mid = len(arr) // 2
       left_half = merge_sort(arr[:mid])
       right_half = merge_sort(arr[mid:])
       return merge(left_half, right_half)

   def merge(left, right):
       result = []
       i = j = 0
       while i < len(left) and j < len(right):
           if left[i] < right[j]:
               result.append(left[i])
               i += 1
           else:
               result.append(right[j])
               j += 1
       result.extend(left[i:])
       result.extend(right[j:])
       return result
   ```

5. **O(n^2) - Quadratic Time Complexity:**
   - Efficiency grows with the square of the input size.

   ```python
   def quadratic_example(matrix):
       rows, cols = len(matrix), len(matrix[0])
       for i in range(rows):
           for j in range(cols):
               print(matrix[i][j])
   ```
